```python
from numpy import arange
from matplotlib import pylab as plt
%matplotlib inline
plt.style.use('ggplot')
```

```python
#Section 3.1 of Faul and Jackson (2005, EPSL)

def PtoZ(Pgpa):
    from numpy import exp
    P = Pgpa * 1e9
    g=9.8
    beta=8.7e-12
    rho0 = 3350.
    num = 1. - exp(-P*beta)
    den = g*beta*rho0
    return num/den / 1000.0

ps = arange(15)
_ = plt.plot(ps,PtoZ(ps),lw=3,label='Faul and Jackson (2005)')
_ = plt.plot(ps,ps*30,lw=3, label='1 GPa = 30 km')
_ = plt.ylim(450,0)
_,_=plt.ylabel('Depth (km)'),plt.xlabel('Pressure (GPa)')
_ = plt.legend()
```
